```python
def encoding_cancer(df, column, name):
    converted = []
    for index in range(len(df)):
        if df.iloc[index, column] == 'cancer':
            converted.append(1)
       # elif df.iloc[index, column] == 0: #this separates task out, but not sure I like it...
           # converted.append(1)
        else:
            converted.append(0)
    df[name] = converted
    return df
```

## Encoding remaining data

```python
def encoding(df, column, name):
    converted = []
    for index in range(len(df)):
        if df.iloc[index, column] == 'Yes' or df.iloc[index, column] == 'Public' or df.iloc[index, column] == 'svm' or df.iloc[index, column] == 'accuracy':
            converted.append(1)
        elif df.iloc[index, column] == 'No' or df.iloc[index, column] == 'Private'or df.iloc[index, column] == 'other':
            converted.append(0)
        elif df.iloc[index, column] == 'Mixed' or df.iloc[index, column] == 'neural network':
            converted.append(2)
    df[name] = converted
    return df
```

```python
df = encoding(df2, 2, 'dataset type encoded')
df = encoding(df, 3, 'demographics encoded')
df = encoding(df, 4, 'affiliations encoded')
df = encoding(df, 5, 'ethics encoded')
df = encoding(df, 6, 'method encoded')
df = encoding(df, 7, 'measure encoded')
```
