```python
import re
import pandas as pd
import numpy as np
```

```python
def find_abstract(year, place):
        abstract_check = []
        abstract_indicator = []
        abstract_list = []
        with open(place, "r", encoding = 'utf-8') as part:
            article = part.read()           
            abstracts = [i.start() for i in re.finditer("Abstract", article)]
            for index in abstracts:
                
                if year == 2012 and article[index+8] != "s": #removing the "Abstracts", aka the references in 2012
                    abstract_list.append(article[index:index+2000])
                    if article[index:index+2000].find("classification") > 0 or article[index:index+2000].find("prediction") > 0 or article[index:index+2000].find("detection") > 0 or article[index:index+2000].find("predict") > 0 or article[index:index+2000].find("classifier") > 0:
    
                        abstract_indicator.append(1)
                    else:
                        abstract_indicator.append(0)
                
                elif year == 2021 and article[index+9] != "T": #removing the references in 2021 ("abstract track")
                    abstract_check.append(article[index-8000:index+100])#used to find and weed out wrong hits
                    abstract_list.append(article[index:index+2000])
                    if article[index:index+2000].find("classification") > 0 or article[index:index+2000].find("prediction") > 0 or article[index:index+2000].find("detection") > 0 or article[index:index+2000].find("predict") > 0 or article[index:index+2000].find("classifier") > 0:
    
                        abstract_indicator.append(1)
                    else:
                        abstract_indicator.append(0)          
       
            
        return abstract_list, abstract_indicator, abstract_check
```

```python
abstract_list2012, abstract_indicator2012, abstract_check2012 = find_abstract(2012, r"C:\Users\chris\Desktop\Value-Analysis-Thesis\combining proceedings txt\miccai_2012_full_txt.txt")
abstract_list2021, abstract_indicator2021, abstract_check2021 = find_abstract(2021, r"C:\Users\chris\Desktop\Value-Analysis-Thesis\combining proceedings txt\miccai_2021_full_txt.txt")
```

```python
#reading in df
df_2012 = pd.read_csv('database_miccai_2012_first_wave.csv')
df_2021 = pd.read_csv('database_miccai_2021_first_wave.csv')
```

```python
#adding column: classification indicated
def classification_indicated(df):
    class_list = []
    for i in range(len(df)):
        if df.loc[i,'Keywords in title'] == 1 and df.loc[i,'Abstract scanned'] == 1:
            class_list.append(1)
        else:
            class_list.append(0)
    df['Classification indicated'] = class_list
    return df
```

```python
classification_indicated(df_2012)
classification_indicated(df_2021)
```
